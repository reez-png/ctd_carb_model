# Step 02 — Fit and inspect models

Fit MLR (and ANN) on the measured responses and look at the MLR coefficients (interpretable: which CTD variable drives pH/TA, and in what direction).

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import pandas as pd
```

```python
# --- parameters ---
MODEL_READY = '../outputs/model_ready.csv'
```

```python
from ctd_carb_model import data, models
ds = data.prepare(pd.read_csv(MODEL_READY))
measured = [r for r in ds.responses if ds.response_status[r]=='measured']
for resp in measured:
    m = models.build_model('MLR', ds.predictors).fit(ds.frame, resp)
    print(resp, 'MLR coefficients:')
    for k,v in m.coefficients.items(): print(f'   {k:20s} {v:+.5g}')
    print()
```

Coefficient signs should be physically sensible (e.g. pH falling with depth, TA rising with salinity).
