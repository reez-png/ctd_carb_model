# Step 06 — Generate predicted carbonate profiles

Train the MLR on the bottle table, apply it to the FULL CTD profile to predict pH/TA at every level, recompute DIC/pCO2/omega from the predicted pair, and flag extrapolation (casts with no local carbonate training).

⚠️ With a weak model (low out-of-group R²), these are **exploratory** predicted fields, clearly labelled — not validated carbonate profiles. The output feeds the plotting tool's transect sections.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import pandas as pd
```

```python
# --- parameters ---
BOTTLE = r'C:\Users\OA_2023-03\Projects\oxytest\data\merged_ctd_carbonate.csv'
PROFILE = r'C:\Users\OA_2023-03\Projects\oxytest\data\merged_ctd_carbonate_full_profile.csv'
OUT = r'..\outputs\predicted_profiles.csv'
SELECT_BY = 'aicc'   # AICc-select predictors per response
```

```python
from ctd_carb_model import predict
bottle = pd.read_csv(BOTTLE, low_memory=False)
profile = pd.read_csv(PROFILE, low_memory=False)
pred, info = predict.predict_profiles(bottle, profile, select_by=SELECT_BY)
print(predict.summarise(info))
predict.write_predicted_profiles(pred, OUT)
print('\nwrote', OUT)
```
