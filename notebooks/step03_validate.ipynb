# Step 03 — Leakage-safe validation

Leave-one-cast-out: hold out a whole cast, predict it, repeat. The out-of-group R²/RMSE here is the honest measure of skill. Ignore in-sample fit.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import pandas as pd
```

```python
# --- parameters ---
MODEL_READY = '../outputs/model_ready.csv'
SPLIT = 'leave_one_cast_out'
```

```python
from ctd_carb_model import data, models, validation
ds = data.prepare(pd.read_csv(MODEL_READY))
print(validation.readiness(ds.frame)); print()
measured = [r for r in ds.responses if ds.response_status[r]=='measured']
for resp in measured:
    cv = validation.cross_validate(ds.frame, lambda: models.build_model('MLR', ds.predictors), resp, SPLIT)
    if cv['ok']:
        p=cv['pooled']; print(f"{resp}: out-of-group R2={p['r2']:.3f}, RMSE={p['rmse']:.4g} ({cv['n_folds']} folds)")
    else:
        print(f"{resp}: {cv['note']}")
```

A negative R² means worse than predicting the mean. The readiness note tells you what you can and cannot claim with your current coverage.
