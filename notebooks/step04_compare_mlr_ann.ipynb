# Step 04 — Compare MLR vs ANN

Both models, same folds, same metrics, one table, with a reasoned winner per response. On a small dataset, expect MLR to win and the ANN to struggle — the comparison shows it honestly.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import pandas as pd
```

```python
# --- parameters ---
MODEL_READY = '../outputs/model_ready.csv'
SPLIT = 'leave_one_cast_out'
SELECT_BY = 'aicc'   # 'aicc' to AICc-select predictors first, or None to use all
```

```python
from ctd_carb_model import data, compare
ds = data.prepare(pd.read_csv(MODEL_READY))
measured = [r for r in ds.responses if ds.response_status[r]=='measured']
result = compare.compare_models(ds.frame, ds.predictors, measured, ['MLR','ANN'], SPLIT, select_by=SELECT_BY)
print(compare.format_comparison(result))
```

The winner declaration includes the reason. If MLR and ANN are close, the pipeline prefers MLR (simpler, safer on small data).
