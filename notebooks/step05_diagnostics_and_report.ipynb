# Step 05 — Diagnostics & report

The convincing artifacts: predicted-vs-actual plots, residuals, per-fold panels, comparison bars, a markdown report, and a manifest. All from out-of-group predictions.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
import pandas as pd
```

```python
# --- parameters ---
MODEL_READY = '../outputs/model_ready.csv'
SPLIT = 'leave_one_cast_out'
OUT_DIR = '../outputs/model_products'
```

```python
from ctd_carb_model import data, compare, report
ds = data.prepare(pd.read_csv(MODEL_READY))
measured = [r for r in ds.responses if ds.response_status[r]=='measured']
result = compare.compare_models(ds.frame, ds.predictors, measured, ['MLR','ANN'], SPLIT)
written = report.generate_all(result, OUT_DIR, ds.response_status)
report.write_report(result, OUT_DIR + '/report.md', data.summarise(ds), ds.response_status)
report.write_manifest(result, OUT_DIR + '/manifest.json', {'master': MODEL_READY}, data.summarise(ds))
print(f"{len(written['plots'])} plots + report + manifest written to {OUT_DIR}")
for p in written['plots']: print('  ', p)
```
